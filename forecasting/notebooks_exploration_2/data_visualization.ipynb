```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import make_pipeline
from sklearn.cluster import DBSCAN
```

```python
tickers = ['OBNK','UEIC','FERG','SPUC','CPRI']
df = pd.read_pickle(f'./{tickers}_daily.pkl')
display(df)
# display(df.columns.unique())
dfn = df.dropna()
```

```python

pipe = make_pipeline(MinMaxScaler(),PCA(n_components=0.95))
pipe.fit(dfn)
```

```python
pca = pipe['pca']
# display(pca.components_)
# display(pca.explained_variance_)
display(pca.explained_variance_ratio_)
# display(pca.noise_variance_)
```

```python
pipe = make_pipeline(MinMaxScaler(),DBSCAN(n_jobs=-1))
pipe.fit(dfn)
```
