```python
df = pd.read_pickle("['AAPL']_daily.pkl")
display(df)
display(df.info())
display(df.describe)
```

```python
windl = df.AAPL.close.rolling(window=3, min_periods=1, closed='left')
means = windl.mean()
windr = df.AAPL.close.rolling(window=3, min_periods=1, closed='right')
rights = windr.mean()
windc = df.AAPL.close.rolling(window=1, min_periods=1, closed='both')
c = windc.mean()
targets = pd.DataFrame({'lefts':means, 'rights':rights, 'c':c})
targets
```
